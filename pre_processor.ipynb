```python
col_desc_index = {}
col_descricao_index = {}

for i, col in enumerate(col_names):
    linha = dic[dic["Código_da_Váriavel"] == col]
    
    if not linha.empty:
        descricao = linha["Quesito_Descrição"].values[0]
        col_desc_index[col] = [i, descricao]
    else:
        col_desc_index[col] = [i, None]

col_desc_index
```

```python
import pandas as pd
cols_ausentes = {}
cols_ausentes_list = []
def contar_nulos(df: pd.DataFrame):
    nulos = df.isnull().sum()
    nulos = nulos[nulos > 0]
    return nulos.to_dict()


dff = pd.read_csv("./data/df_filter_90k.csv")
nulos_por_coluna = contar_nulos(dff)

for coluna, n_nulos in nulos_por_coluna.items():
    # print(f"{coluna}: {n_nulos} valores nulos")
    cols_ausentes_list.append(coluna)
    cols_ausentes[coluna] = n_nulos
    
ordenado_decrescente = dict(sorted(cols_ausentes.items(), key=lambda item: item[1], reverse=True))
print(ordenado_decrescente)
```
