Colunas que ainda possuem valores nulos e a quantidade

```python
import pandas as pd
cols_ausentes = {}
cols_ausentes_list = []
def contar_nulos(df: pd.DataFrame):
    nulos = df.isnull().sum()
    nulos = nulos[nulos > 0]
    return nulos.to_dict()


dff = pd.read_csv("./data/df_filter_90k_jan_version.csv")
nulos_por_coluna = contar_nulos(dff)

for coluna, n_nulos in nulos_por_coluna.items():
    # print(f"{coluna}: {n_nulos} valores nulos")
    cols_ausentes_list.append(coluna)
    cols_ausentes[coluna] = n_nulos
    
ordenado_decrescente = dict(sorted(cols_ausentes.items(), key=lambda item: item[1], reverse=True))
print(ordenado_decrescente)
```

```python
labels = [each for each in every_values.keys()]
print(labels)
values = [int(each) for each in every_values.values()]
print(values)

plt.bar(labels, values)
plt.title(f"Qtd de elementos da várivels target: {target_col}")
plt.ylabel("Qtd de registros")
plt.xlabel("Categoria dos registros")
```
